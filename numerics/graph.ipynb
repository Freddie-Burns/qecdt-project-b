```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
G.add_nodes_from([1, 2, 3, 4])
G.add_edge(1, 2)
G.add_edge(2, 3)
G.add_edge(3, 1)
G.add_edge(3, 1)
```

```python
nx.draw_circular(G, with_labels=True, font_weight='bold')
```

```python
G = nx.complete_graph(6)
nx.draw_circular(G, with_labels=True, font_weight='bold')
```
